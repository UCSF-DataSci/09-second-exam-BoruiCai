```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Walking speed analysis:

```python
df = pd.read_csv('question2.csv')

#Scatter plot of age vs walking speed with regression line
plt.figure(figsize=(10, 6))
sns.lmplot(x="age", y="walking_speed", data=df)
plt.title("Scatter Plot of Age vs Walking Speed")
plt.xlabel("age")
plt.ylabel("Walking Speed")
plt.show()
```

```python
## Box Plot: Walking Speed by Education Level
plt.figure(figsize=(10, 6))
sns.boxplot(x="education_level", y="walking_speed", data=df)
plt.title("Walking Speed by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Walking Speed")
plt.show()
```

```python
## Education-Age Interaction line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="age", y="walking_speed", hue="education_level", data=df)
plt.title("Line Plot of Education-Age Interaction on Walking Speed")
plt.show()
```

## Cost analysis:

```python
# Bar plot of mean costs by insurance type
plt.figure(figsize=(10, 6))
sns.barplot(x="insurance_type", y="visit_cost", data=df, errorbar='sd')
plt.title("Bar plot of mean costs by insurance type")
plt.xlabel("Insurance Type")
plt.ylabel("Mean Visit Cost")
plt.show()
```

```python
# Box plots showing cost distributions
plt.figure(figsize=(10, 6))
sns.boxplot(x="insurance_type", y="visit_cost", data=df)
plt.title("Box Plot of Cost Distributions by Insurance Type")
plt.xlabel("Insurance Type")
plt.ylabel("Visit Cost")
plt.show()
```

## Combined visualizations

```python
# Pair plot of key variables
pairplot = sns.pairplot(
    df[["age", "walking_speed", "visit_cost", "insurance_type"]], hue="insurance_type",palette="tab10")
plt.show()
```

```python
# Faceted Scatter Plotfor  Education Level
g = sns.FacetGrid(df, col="education_level",  hue="education_level", palette="tab10")
g.map(sns.scatterplot, "age", "walking_speed")
g.add_legend()
g.savefig("facetgrid_education_level_colored.png")
plt.show()

# Faceted Scatter Plotfor  insurance type
c = sns.FacetGrid(df, col="insurance_type",  hue="insurance_type", palette="tab10")
c.map(sns.scatterplot, "age", "walking_speed")
c.add_legend()
c.savefig("facetgrid__insurance_type_colored.png")
plt.show()
```

```python
# Plot: Visit Cost Trends Over Time
plt.figure(figsize=(10, 6))
sns.lineplot(x="visit_date", y="visit_cost", hue="insurance_type", data=df, ci=None)
plt.title("Visit Cost Trends Over Time by Insurance Type")
plt.legend(title="Insurance Type")
plt.savefig("visit_cost_time_trend.png")
plt.show()
```
